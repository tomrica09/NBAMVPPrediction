##### Cleaning the Player Data

```python
import pandas as pd 

players = pd.read_csv("[2024-2025]_players.csv")
players
```

```python
# Fetch the row with index 62 (checking for the repeated columns with header labels data)
row_at_index_62 = players.iloc[60]
row_at_index_62
```

```python
# Filtering out the repeated columns with header labels data
players = players[players["Player"] != "Player"]
```

```python
# Delete unnecessary columns
del players["Unnamed: 0"]
del players["Rk"]
del players["Awards"]
```

```python
# Removing the asterisk on some player stats (for assurance only)
players.loc[:, "Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
# Change "Team" column to "Tm"
players.rename(columns={'Team': 'Tm'}, inplace=True)
players
```

```python
# Filtering out the multiple entries/teams of the players in a single year and remain only the last team they played for in that year
def single_team(df):
    if df.shape[0] == 1:
        return df
    else:
        # Correctly filter using .isin() for multiple conditions
        row = df[df["Tm"].isin(["2TM", "3TM", "4TM"])]
        row["Tm"] = df.iloc[-1, :]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
# Drop the multi-level index (first index which is the player name)
players.index = players.index.droplevel()
```

```python
# Drop the multi-level index (second index which is the year)
players.index = players.index.droplevel()
```

##### Cleaning the Team Data

```python
teams = pd.read_csv("[2024-2025]_teams.csv")
teams
```

```python
# Getting rid of Division Rows
teams = teams[~teams["W"].str.contains("Division")]
teams
```

```python
# Remove asterisk on team names
teams.loc[:, "Team"] = teams["Team"].str.replace("*", "", regex=False)
teams
```

```python
import re

# Function to clean team names
def clean_team_name(team_name):
    return re.sub(r'\s*\(.*?\)\s*', '', team_name)

# Apply the function to the 'Team' column
teams['Team'] = teams['Team'].apply(clean_team_name)

# Show the cleaned DataFrame
teams
```

```python
# Delete unnecessary columns
del teams["Unnamed: 0"]
```

```python
teams
```

```python
# Integrate the team names into the players DataFrame
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]: # Skipping the first row
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name

# Show the nicknames dictionary
nicknames
```

```python
# Integrate the team names into the players DataFrame
players["Team"] = players["Tm"].map(nicknames)
players.head(5)
```

```python
stats = players.merge(teams, how="outer", on=["Team", "Year"])
stats
```

```python
stats.dtypes
```

```python
# Convert the numerical columns to numeric
stats = stats.apply(pd.to_numeric, errors="ignore")
stats.dtypes
```

```python
stats["GB"].unique()
```

```python
# Fill the dashes with 0
stats["GB"] = stats["GB"].str.replace("—", "0")
stats["GB"].unique()
```

```python
# Converting GB column to numeric
stats["GB"] = pd.to_numeric(stats["GB"])
stats.dtypes
```

```python
# Storing the merged player and team stats
stats.to_csv("[2024-2025]_player_team_stats.csv")
```
