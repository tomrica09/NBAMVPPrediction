##### Cleaning the Player Data

```python
players = pd.read_csv("players.csv")
players
```

```python
# Fetch the row with index 62 (checking for the repeated columns with header labels data)
row_at_index_62 = players.iloc[60]
row_at_index_62
```

```python
# Filtering out the repeated columns with header labels data
players = players[players["Player"] != "Player"]
```

```python
players.iloc[90]
```

```python
# Delete unnecessary columns
del players["Unnamed: 0"]
del players["Rk"]
del players["Awards"]
```

```python
players
```

```python
# Removing the asterisk on some player stats (for assurance only)
players.loc[:, "Player"] = players["Player"].str.replace("*", "", regex=False)
```

```python
players.groupby(["Player", "Year"])
```

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991))
```

```python
# Change "Team" column to "Tm" to match it to mvps dataframe
players = players.rename(columns={'Team': 'Tm'})
```

```python
players
```

```python
# Filtering out the multiple entries/teams of the players in a single year and remain only the last team they played for in that year
def single_team(df):
    if df.shape[0] == 1:
        return df
    else:
        # Correctly filter using .isin() for multiple conditions
        row = df[df["Tm"].isin(["2TM", "3TM", "4TM"])]
        row["Tm"] = df.iloc[-1, :]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players.head(20)
```

```python
# Drop the multi-level index (first index which is the player name)
players.index = players.index.droplevel()
```

```python
# Drop the multi-level index (second index which is the year)
players.index = players.index.droplevel()
```

```python
players
```

```python
players[players["Player"] == "Greg Anderson"]
```

##### Combining the Player and MVP Data

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])
```

```python
combined.head(5)
```

```python
combined[combined["Pts Won"] > 0]
```

```python
# Changin NaN values to 0
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)
```

```python
combined
```

##### Cleaning the Team Data

```python
teams = pd.read_csv("teams.csv")
teams
```

```python
teams.head(30)
```

```python
# Getting rid of Division Rows
teams = teams[~teams["W"].str.contains("Division")]
```

```python
teams.head(30)
```

```python
# Remove asterisk on team names
teams.loc[:, "Team"] = teams["Team"].str.replace("*", "", regex=False)
```

```python
teams.head(5)
```

```python
teams["Team"].unique()
```

```python
combined["Tm"].unique()
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]: # Skipping the first row (header)
        abbrev,name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
nicknames
```

```python
combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head(5)
```

```python
stats = combined.merge(teams, how="outer", on=["Team", "Year"])
```

```python
stats
```

```python
del stats["Unnamed: 0"]
```

```python
stats = stats.apply(pd.to_numeric, errors="ignore")
```
